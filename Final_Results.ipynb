```python
import pandas as pd
import re

# Read the CSV file
df = pd.read_csv('train.csv')

# Function to clean the text (remove only asterisks)
def clean_text(text):
    if isinstance(text, str):
        # Remove asterisks (*) only
        cleaned = re.sub(r'\*', '', text)
        # Remove extra spaces
        cleaned = re.sub(r'\s+', ' ', cleaned).strip()
        return cleaned
    return text

# Apply the cleaning function to the 'Translated_DealerComment' column
df['DealerComment'] = df['Translated_DealerComment'].apply(clean_text)

# Save the cleaned data to a new CSV file
df.to_csv('train.csv', index=False)
```

Exploratory Data Analysis in train data

```python
#1. import the libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load the dataset
file_path = 'train.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

#3. List of columns to analyze
analyze = ['Component', 'DefectivePart', 'Failure', 
                      'PartsReplaced', 'Cost']

#4. Loop through each column and plot its distribution
for column in analyze:
    plt.figure(figsize=(10, 6))
    # Count occurrences of each value in the column
    value_counts = df[column].value_counts()
    
    # Plot the distribution
    value_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Number of Occurrences', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    
    # Show each plot
    plt.show()
```

Using Google translator for dealer comment column

```python
import pandas as pd
from googletrans import Translator

# Read the CSV file
df = pd.read_csv('test_17_New.csv')

# Check the column names to ensure 'Dealercomment' exists
print("Column names:", df.columns)

# Initialize the translator
translator = Translator()

# Function to translate text
def translate_text(text):
    try:
        return translator.translate(text, dest='en').text
    except Exception as e:
        print(f"Error translating text: {text}\nError: {e}")
        return text  # Return original text if translation fails


# Example: if the column name is 'DealerComment' instead of 'Dealercomment', adjust the code accordingly
if 'DealerComment' in df.columns:  # Use correct case for the column
    df['Translated_DealerComment'] = df['DealerComment'].apply(translate_text)
else:
    print("The 'DealerComment' column does not exist. Please check the column names.")
    
# Save the translated results into a new CSV file
df.to_csv('test.csv', index=False)

print("Translation complete. Saved to 'test.csv'")
```

```python
#imort libraries
from transformers import pipeline
import pandas as pd

# Load the dataset
df = pd.read_csv('test.csv')

# Initialize the zero-shot classification using pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define the candidate labels for cause of failure
cause_labels = ["Human error", "Design Flaw", "Material issue"]

# Function to classify the cause of failure
def classify_cause(text):
    result = classifier(text, cause_labels)
    return result['labels'][0]

# Apply the classification to the DealerComment column
df['Cause of Failure'] = df['Translated_DealerComment'].apply(classify_cause)

# Create separate columns for each cause
for cause in cause_labels:
    df[cause] = df['Cause of Failure'].apply(lambda x: 1 if x == cause else 0)

# Display the results
print(df[['Translated_DealerComment', 'Cause of Failure', 'Human error', 'Design Flaw', 'Material issue']].head())

# Calculate the distribution of causes
cause_distribution = df['Cause of Failure'].value_counts(normalize=True)
print("\nDistribution of Causes of Failure:")
print(cause_distribution)                               
```
